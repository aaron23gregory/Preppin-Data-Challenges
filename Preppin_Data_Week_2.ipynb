For week 2 of our beginner month, Data Source Bank has a requirement to construct International Bank Account Numbers (IBANs), even for Transactions taking place in the UK. We have all the information in separate fields, we just need to put it altogether in the following order:





Inputs
A list of the transactions, with information about the receiving bank account 


A lookup table for the SWIFT Bank Codes 


Requirements

Input the data
In the Transactions table, there is a Sort Code field which contains dashes. 
We need to remove these so just have a 6 digit string (hint)

Use the SWIFT Bank Code lookup table to bring in additional information about the SWIFT code and Check Digits of the receiving bank account (hint)

Add a field for the Country Code (hint)
Hint: all these transactions take place in the UK so the Country Code should be GB

Create the IBAN as above (hint)
Hint: watch out for trying to combine sting fields with numeric 
fields - check data types

Remove unnecessary fields (hint)

Output the data

```python
from google.colab import files
import io
import pandas as pd

uploaded = files.upload()
```

```python
import pandas as pd
transactions = pd.read_csv(io.BytesIO(uploaded['Transactions.csv']))
transactions.head() 
```

```python
uploaded = files.upload()
swift = pd.read_csv(io.BytesIO(uploaded['Swift Codes.csv']))
swift.head() 
```

```python
#Remove dash marks from the sort code column in the transaction table
transactions['Sort Code'] = transactions['Sort Code'].str.replace('-', '')
transactions.head()
```

```python
#Use the SWIFT Bank Code lookup table to bring in additional information about the SWIFT code and Check Digits 
#of the receiving bank account
swift['Check Digits'] = swift['Check Digits'].astype(str)
merged_df = pd.merge(transactions, swift, on='Bank', how='inner')
merged_df.head(5)
```

```python
#Add a field for the Country Code
merged_df["Country Code"]= "GB"
merged_df.head()
```

```python
#Create the IBAN (Country code+ Check Digits+ Bank Code+ Sort Code+ Account Number)
merged_df['IBAN'] = merged_df.apply(lambda row: ''.join([str(row['Country Code']), str(row['Check Digits']), row['SWIFT code'], 
                                                         str(row['Sort Code']), str(row['Account Number'])]), axis=1)
merged_df.head()
```
