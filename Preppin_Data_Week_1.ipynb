```python
import io
df2 = pd.read_csv(io.BytesIO(uploaded['PD 2023 Wk 1 Input.csv']))
df2.head()  
```

```python
#Split the first part of the transaction code that represents the bank
df2['Bank'] = df2['Transaction Code'].apply(lambda x: x.split("-")[0])
df2.head()
```

```python
#Replacing the 1's and 2's in "Online or in person" with "Online" or "In-Person"
replacement_mapping = {1: 'Online', 2: 'In-Person'}
df2['Online or In-Person'] = df2['Online or In-Person'].map(replacement_mapping)
df2.head()
```

```python
#Change the transaction date to a day of the week
df2['Transaction Date'] = pd.to_datetime(df2['Transaction Date'])

# Extract the day of the week and assign it to a new column
df2['Day of the Week'] = df2['Transaction Date'].dt.strftime('%A')
df2.head()
```

```python
#Grouping value by bank
grouped=df2.groupby("Bank")
print(grouped['Value'].sum())
```

```python
#Grouping by Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)

grouped=df2.groupby(["Bank","Day of the Week","Online or In-Person"])
print(grouped['Value'].sum())
```

```python
#Grouping by Total Values by Bank and customer code

grouped=df2.groupby(["Bank","Customer Code"])
print(grouped['Value'].sum())
```
